■方針
①input1, input2を読み込む。②それぞれ重複を削除する③inut2における１４式竹筒銃についてcategoryの甲をitem_nameに追記したのちに削除、それぞれ左に詰める
④input1, input2を結合する
⑤unit*meanの値を用いて新たにsales列を作成する。

```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
##input1のcsvをインポート
uploaded = files.upload()
```

```python
import io

input1 = pd.read_csv(io.BytesIO(uploaded['input1.csv']))
input1
```

列名に空白があったため名前指定してdfを作成。その後１行目（元のheaderを削除した）

```python
input2_pre = pd.read_csv(io.BytesIO(uploaded['input2.csv']), names = ("ymd", "item_code", "item_name", "category", "unit", "mean", "dummy"))
input2 = input2_pre.drop(0, axis=0)
input2
```

それぞれ重複を削除

```python
input1_dd = input1[~input1.duplicated()]
input1_dd
```

```python
input2_dd = input2[~input2.duplicated()]
input2_dd
```

*③inut2における１４式竹筒銃についてcategoryの甲をitem_nameに追記したのちに削除、それぞれ左に詰める* 

```python
##input2_ddから編集したい行を抽出
input2_edit1 = input2_dd[2:3]
input2_edit1
```

```python
##該当箇所を修正
input2_edit1.loc[3, "item_name"] = "14式竹筒銃甲"
input2_edit1.loc[3, "category"] = "チャージャー"
input2_edit1.loc[3, "unit"] = 14
input2_edit1.loc[3, "mean"] = 150.0
input2_edit1.loc[3, "dummy"] = np.NaN
input2_edit1
```

```python
##input2同士を結合して変更箇所を削除,'item_code', 'unit', 'mean'のデータ型をintに変換
input2_merged = pd.concat([input2_dd, input2_edit1], ignore_index= True)
input2_merged = input2_merged.drop(input2_merged.index[[2]])
input2_merged = input2_merged.drop("dummy", axis=1)
input2_merged['item_code'] = input2_merged['item_code'].astype('int')
input2_merged['unit'] = input2_merged['unit'].astype('int')
input2_merged['mean'] = input2_merged['mean'].astype('int')
input2_merged = input2_merged.sort_values(by='item_code', ascending=True)
input2_merged
```

```python
##加工後のinput1のデータとinput2のデータそれぞれを統合する
output_pre = pd.concat([input1_dd, input2_merged], ignore_index=True)
output_pre
```

```python
##sales列を追加
output_pre['sales'] = output_pre['unit']*output_pre['mean']
output_pre
```

```python
##ymdの表記の変更
output_pre['ymd'] = output_pre['ymd'].str.replace('/', '')
output_pre
```

```python
#各列のデータ型を調整
output_pre['ymd'] = output_pre['ymd'].astype('int')
output_pre['sales'] = output_pre['sales'].astype('float')
output_pre
```

```python
#最終output(下にoutput.csvのデータフレームを記載)※index8のsalesがNaNですが今回はデータ埋めを行いました。
output = output_pre
output
```
